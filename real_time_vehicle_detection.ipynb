```python
# Install necessary libraries
!pip install opencv-python-headless pytesseract --quiet

import cv2
import pytesseract
from google.colab.patches import cv2_imshow
import os
import numpy as np

# Configure Tesseract executable path
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# Upload the video file to Colab
from google.colab import files
uploaded = files.upload()

# Get the video file path
video_path = list(uploaded.keys())[0]

# Load the video
cap = cv2.VideoCapture(video_path)

# Load Haar Cascade for license plate detection
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')

# Create a folder to save processed frames
os.makedirs("frames_with_detections", exist_ok=True)

# Initialize frame counter
frame_count = 0
saved_frames = 0

# Helper function for preprocessing
def preprocess_plate(plate_image):
    gray = cv2.cvtColor(plate_image, cv2.COLOR_BGR2GRAY)
    # Apply Gaussian blur
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    # Enhance contrast
    enhanced = cv2.equalizeHist(blur)
    return enhanced

# Process video and save frames with detected plates
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect number plates
    plates = plate_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(50, 20))
    if len(plates) > 0:
        for (x, y, w, h) in plates:
            # Extract the plate region
            plate_region = frame[y:y + h, x:x + w]

            # Preprocess the plate for OCR
            preprocessed_plate = preprocess_plate(plate_region)

            # Extract text using Tesseract
            plate_text = pytesseract.image_to_string(preprocessed_plate, config='--psm 7')

            # Draw rectangle and text on the frame
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, plate_text.strip(), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

        # Save the frame with detections
        frame_path = f"frames_with_detections/frame_{saved_frames + 1}.jpg"
        cv2.imwrite(frame_path, frame)
        saved_frames += 1

cap.release()
cv2.destroyAllWindows()

print(f"Processed {saved_frames} frames with detected plates and saved them in the 'frames_with_detections' folder.")

# Create a video from the saved frames
frame_files = sorted([os.path.join("frames_with_detections", f) for f in os.listdir("frames_with_detections")])
output_video_path = 'detected_plates_video.avi'

# Get frame dimensions
if len(frame_files) > 0:
    first_frame = cv2.imread(frame_files[0])
    height, width, layers = first_frame.shape

    # Define the codec and create VideoWriter object
    out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'XVID'), 10, (width, height))

    # Write frames into the video
    for frame_file in frame_files:
        frame = cv2.imread(frame_file)
        out.write(frame)

    out.release()
    print(f"Video saved as '{output_video_path}'.")

    # Download the video
    files.download(output_video_path)
else:
    print("No frames with detected plates were saved.")
```
